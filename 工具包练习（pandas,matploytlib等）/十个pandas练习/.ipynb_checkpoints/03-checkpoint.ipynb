```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
drinks = pd.read_csv('exercise_data/drinks.csv')
drinks.head()
```

### step1: 哪个大陆(continent)平均消耗的啤酒(beer)更多？

```python
drinks.groupby('continent')['beer_servings'].mean()
```

### step2: 打印出每个大陆(continent)的红酒消耗(wine_servings)的描述性统计值

```python
drinks.groupby('continent')['beer_servings'].describe()
```

### step3: 打印出每个大陆每种酒类别的消耗平均值¶

```python
drinks.groupby('continent').mean()
# drinks.groupby('continent').median()
```

### step4: 打印出每个大陆对spirit饮品消耗的平均值，最大值和最小值¶

```python
drinks.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min', 'std'])
```
